# Ejercicio 1: Serie de Fourier

La serie de Fourier de nuestra función se expresa de la siguiente manera:
$$f(x)=\frac{1}{2}+\frac{2}{\pi}\sum_{n=1}^{\infty} \left[\frac{1}{n^2 \pi} (1-(-1)^n)+\frac{1}{n}\right].$$

```python
from pylab import *
from math import exp
```

Para empezar, graficamos la condición inicial.

```python
x1 = linspace(-2,0,1000)
x2 = linspace(0,2,1000)
y1=0*x1
y2=2-x2
x=concatenate((x1,x2))
f=concatenate((y1,y2))
plot(x,f,label='$f(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Función a trozos")
xlim([-2.01,2.01]),ylim([-0.01,2.01]),grid(),legend()
show()
```

Usamos una serie de Fourier para aproximar L$.

```python
def u_0(x,N=2): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += (-1)/(n**2*pi)*(1-(-1)**n)+1/n
    return (2.0/pi)*f
```

```python
f_n = 1/2 + u_0(x)
plot(x,f_n,label='$f_N(x)$ con $N=2$')
plot(x,f,label='$f(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.01,2.01]),ylim([0,2.1]),grid(),legend()
show()
```

Podemos visualizar la aproximación para diferentes cantidades de armónicos.

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.01,2.01]),ylim([-0.01,2.1]),grid(),legend()
show()
```

Ahora, se agregará la parte temporal para visualizar la solución completa, comenzando por $t=0$.

```python
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += (-1)/(n**2*pi)*(1-(-1)**n)+1/n
    return (2.0/pi)*f
```

```python
x = linspace(0,1,1000)
t = linspace(0,10,1000)
U = 1/2+u_x_t(x,t)
```

```python
plot(x,U[:,0],label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.01,2.01]),ylim([-0.01,2.1]),grid(),legend()
show()
```

Luego, para diferentes tiempos, lo cual nos permite ver cómo la solución evoluciona cuando $t \to \infty$.

```python
plot(x,U[:,0],label='$u(x,t=0)$') # Mejorar
plot(x,U[:,1],label='$u(x,t=1)$')
plot(x,U[:,2],label='$u(x,t=2)$')
plot(x,U[:,3],label='$u(x,t=3)$')
plot(x,U[:,10],label='$u(x,t=10)$')
plot(x,U[:,20],label='$u(x,t=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$');
plt.title("Evolución de la solución")
xlim([-2.01,2.01]),ylim([-0.01,2.1]),grid(),legend()
show()
```

Y podemos ver esto en una gráfica en 3D.

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
t = linspace(0,1.0,100)

X,T = np.meshgrid(x,t)

U = 1/2+u_x_t(x,t)

plt.xlabel('$t$'); plt.ylabel('$x$');
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="inferno"); #Otros colores:"viridis", "inferno", "plasma"
```
