a) Verificar que el siguiente problema es de Sturm-Liouville:
$$f''(x)+2f'(x)+\lambda f(x)=0$$, $0<x<1$, $f(0)=0$, $f(1)=0$

Solución:
La expresión completa será multiplicada por $e^{2x}$, obteniendo la expresión,
$$e^{2x}f''(x)+2e^{2x}f'(x)+\lambda e^{2x}f(x)=0$$
$$⇒ e^{2x}[f''(x)+2f'(x)]+\lambda e^{2x}f(x)=0$$
$$⇒ [e^{2x}f'(x)]'+\lambda e^{2x}f(x)=0$$;

Donde $p(X)=e^{2x}$, $g(x)=0$, $σ(X)=2^{2x}$, $a=0$, $b=1$, $κ1=1$, $κ2=0$,$κ3=1$,$κ4=0$

Por lo tanto, el problema es de Sturm-Liouville.

Para resolver el problema se aplicarán los siguientes casos


Para $λ=0$ :

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x=sp.symbols('x')
f=sp.Function('f')
ode1=sp.Eq(sp.Derivative(f(x),x,x)+2*sp.Derivative(f(x),x),0)

sol1 = sp.dsolve(ode1,f(x))

const=sp.solve([sol1.rhs.subs(x,0)-0, sol1.rhs.subs(x,1)-0])
C1, C2 = sp.symbols('C1,C2')
sol1=sol1.subs(const)

func1 = sp.lambdify(x,sol1.rhs,'numpy')
xx = np.arange(0,10,.1)
yy1 = func1(xx) * np.ones_like(xx)

plt.plot(xx,yy1,'r')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
