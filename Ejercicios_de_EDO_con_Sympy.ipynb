Resolver y graficar los siguientes problemas:

1) $X''+ kX=0$, $X(0)=X(1)=0$

2) $X''+ kX=0$, $X'(0)=X'(1)=0$

Solución: Primero se evaluará la ecuación $X''+kX=0$ con $X(0)=X(1)=0$

Para $k=0$

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x=sp.symbols('x')
X=sp.Function('X')
ode11=sp.Eq(sp.Derivative(X(x),x,x),0)

sol11 = sp.dsolve(ode11,X(x))

const=sp.solve([sol11.rhs.subs(x,0)-0, sol11.rhs.subs(x,1)-0])
C11, C21 = sp.symbols('C11,C21')
sol11=sol11.subs(const)

func1 = sp.lambdify(x,sol11.rhs,'numpy')
xx = np.arange(0,10,.1)
yy11 = func1(xx) * np.ones_like(xx)

plt.plot(xx,yy11,'r')
plt.xlabel('x')
plt.ylabel('X(x)')
plt.show()
```

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x=sp.symbols('x')
X=sp.Function('X')
ode12=sp.Eq(sp.Derivative(X(x),x,x)+x*X(x),0)

sol12 = sp.dsolve(ode12,X(x))

const=sp.solve([sol12.rhs.subs(x,0)-0, sol12.rhs.subs(x,1)-0])
C12, C22 = sp.symbols('C12,C22')
sol12=sol12.subs(const)

func2 = sp.lambdify(x,sol12.rhs,'numpy')
xx = np.arange(0,10,.1)
yy12 = func2(xx) * np.ones_like(xx)

plt.plot(xx,yy12,'r')
plt.xlabel('x')
plt.ylabel('X(x)')
plt.show()
```

Posteriormente se evaluará la ecuación $X''+kX=0$ con $X'(0)=X'(1)=0$

Para $k=0$

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x=sp.symbols('x')
X=sp.Function('X')
ode21=sp.Eq(sp.Derivative(X(x),x,x),0)

sol21 = sp.dsolve(ode21,X(x))

sol21 = sol21.subs({sp.symbols('C1'): 1, sp.symbols('C2'): 0})

func4 = sp.lambdify(x,sol21.rhs,'numpy')
xx = np.arange(0,10,.1)
yy21 = func4(xx) * np.ones_like(xx)

plt.plot(xx,yy21,'r')
plt.xlabel('x')
plt.ylabel('X(x)')
plt.show()
```
